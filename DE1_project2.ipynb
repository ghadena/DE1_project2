```python
def count_observations_for_year(df, year, table_name):
    observations = df[df['Year'] == year]
    num_observations = observations.shape[0]
    print(f"Number of observations for the year {year} in {table_name}: {num_observations}")

# Example usage for multiple tables
count_observations_for_year(df_country, 2022, 'country')
count_observations_for_year(df_spending, 2022, 'spending')
count_observations_for_year(df_outcomes, 2022, 'country')
# Repeat for each table as needed
```

```python
count_observations_for_year(df_country, 2019, 'country')
count_observations_for_year(df_spending, 2019, 'spending')
count_observations_for_year(df_outcomes, 2019, 'country')
```

```python
import pandas as pd

# Define the range of years and the expected number of countries
years = range(2009, 2023)
expected_count = 36

# Function to count observations per year in each DataFrame
def count_observations_by_year(df, df_name):
    # Filter for the years in the specified range
    year_counts = df[df['Year'].isin(years)]['Year'].value_counts()
    year_counts = year_counts.reindex(years, fill_value=0)  # Ensure all years are represented, even if count is 0
    return year_counts.rename(df_name)

# Count observations per year for each table
count_country = count_observations_by_year(df_country, 'Country')
count_spending = count_observations_by_year(df_spending, 'Spending')
count_outcomes = count_observations_by_year(df_outcomes, 'Outcomes')

# Combine the counts into a single DataFrame for easy comparison
combined_counts = pd.DataFrame({'Country': count_country, 'Spending': count_spending, 'Outcomes': count_outcomes})

# Filter each column for years with fewer than 36 observations
years_with_fewer_than_36 = combined_counts[(combined_counts < expected_count) & (combined_counts > 0)]

# Display the years with fewer than 36 observations in each table
print("Years with fewer than 36 observations in each table:")
print(years_with_fewer_than_36.dropna(how='all'))  # Drop rows where all values are NaN (no shortages)
```
